## 01.1. Importação das bibliotecas

### 01.1.1. Importando as bibliotecas

```python
import sys

import joblib as jb
import pandas as pd

sys.path.append('..')
import params.consts as consts
```

## 01.2. Deploy

### 01.2.1. Carregando o modelo

```python
model_classification = jb.load(consts.MODEL_CLASSIFICATION_JOBLIB)
```

### 01.2.2. Lendo o dataset a ser previsto

```python
df = pd.read_csv(consts.DATASET_DEPLOY_CLASSIFICATION, sep = ';')
```

### 01.2.3. Fazendo as previsões com o modelo

```python
predictions = model_classification.predict(df)
```

### 01.2.4. Criando uma coluna para armazenar as previsões

```python
df['Response'] = predictions
```

### 01.2.5. Salvando o dataset com as previsões

```python
df.to_csv(consts.DATASET_DEPLOYED_CLASSIFICATION, index = False)
```
